#### IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
devl=pd.read_csv('train.csv')
devl.head(8)
```

```python
devl.isnull()
```

```python
devl.isnull().sum()
```

#### HEATMAP

```python
sns.heatmap(devl.isnull())
```

LINE PLOT

```python
sns.lineplot(data=devl,x="Age",y="Survived")
plt.show()
```

#### COUNT PLOT

```python
sns.countplot(x='SibSp',data=devl,palette='Blues_r')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Sex',data=devl,palette='BuGn')
```

```python
sns.set_style('darkgrid')
sns.countplot(x='Survived',hue='Pclass',data=devl,palette='cubehelix')
```

#### REDPLOT

```python
sns.catplot(x="Pclass",y="Age", data=devl,kind="swarm")
plt.show()
```

#### SCATTER PLOT

```python
sns.pairplot(devl)
```

```python
sns.catplot(data=devl,x="Age",y="Pclass")
plt.title("CATPLOT")
plt.show()
```

#### BAR PLOT

```python
sns.barplot(data=devl,x='Age',y='Survived')
```

```python
sns.barplot(x="Pclass",y="Age", data=devl)
plt.show()
```

#### RELPLOT

```python
sns.relplot(x="Pclass",y="Age", data=devl,kind="line",hue="Sex")
plt.show()
```

#### BOX PLOT

```python
sns.catplot(x="Pclass",y="Age", data=devl,kind="box")
plt.show()
```

#### DISTPLOT

```python
sns.distplot(devl['Age'].dropna(),color='darkviolet',bins=10)
plt.show()
```
